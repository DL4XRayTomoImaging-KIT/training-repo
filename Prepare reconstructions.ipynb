```python
from reco_helper import reconstruct_one_slice
```

```python
import numpy as np
import tifffile
from tqdm.auto import tqdm
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

## 681

```python
data_links = []
for i in tqdm(range(120)):
    data_links.append(np.load(f'/mnt/HD-LSDF/sd18a004/for_yaroslav/projections_681/proj_{i}.npy', mmap_mode='r'))
```

```python
plt.imshow(data_links[0][:, 200])
data_links[0].shape
```

```python
for i in tqdm(range(data_links[0].shape[1]), desc='slices'):
    projections = [s[:, i] for s in tqdm(data_links, leave=False, desc='loading slices')]
    projections = np.stack(projections, 1)
    reconstructed_slice = reconstruct_one_slice(projections)
    tifffile.imwrite(f'/mnt/HD-LSDF/sd18a004/for_yaroslav/reconstructions_681/{i}.tif', reconstructed_slice)
```

```python
img = tifffile.imread('/mnt/HD-LSDF/sd18a004/for_yaroslav/reconstructions_681/0.tif')
plt.imshow(img[200])
```
