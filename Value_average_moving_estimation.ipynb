```python
import tifffile

from matplotlib import pyplot as plt
%matplotlib inline

import os
```

```python
addr_raw_001 = os.path.join(base_folder, '001', 'scan_0001', 'ffcorr_corr_dead_px', 'scan_0001.tif')
addr_den_001 = os.path.join(base_folder, '001', 'scan_0001', 'ffcorr_corr_dead_px', 'denoised_dc_scan_0001.tif')
addr_pag_001 = os.path.join(base_folder, '001', 'pr-pag-all', 'scan_0001-pr-pag-2p0.tif')

addr_raw_036  = os.path.join(base_folder, '036', 'scan_0001', 'ffcorr_corr_dead_px', 'scan_0001.tif')
addr_pag_036  = os.path.join(base_folder, '036', 'scan_0001', 'pr-pag', 'pr-pag-2p0.tif')
```

```python
%%time
raw_001 = tifffile.imread(addr_raw_001)
den_001 = tifffile.imread(addr_den_001)
pag_001 = tifffile.imread(addr_pag_001)

raw_036 = tifffile.imread(addr_raw_036)
pag_036 = tifffile.imread(addr_pag_036)
```

```python
t = 700

fig, axes = plt.subplots(ncols=2, nrows=2, figsize=(10, 10))
axes[0][0].scatter(raw_001[:t, -100:, -100].flatten(), raw_001[1:t+1, -100:, -100].flatten())
axes[0][0].set_title('raw 001/scan_0001')
axes[0][0].set_ylabel('pixel value at time i+1')

axes[0][1].scatter(pag_001[:t, -100:, -100].flatten(), pag_001[1:t+1, -100:, -100].flatten())
axes[0][1].set_title('Paganin 001/scan_0001')

axes[1][0].scatter(raw_036[:t, -100:, -100].flatten(), raw_036[1:t+1, -100:, -100].flatten())
axes[1][0].set_title('raw 036/scan_0001')
axes[1][0].set_xlabel('pixel value at time i')
axes[1][0].set_ylabel('pixel value at time i+1')

axes[1][1].scatter(pag_036[:t, -100:, -100].flatten(), pag_036[1:t+1, -100:, -100].flatten())
axes[1][1].set_title('Paganin 036/scan_0001')
axes[1][1].set_xlabel('pixel value at time i')
```

```python
plt.figure(figsize=(5, 5))
plt.scatter(den_001[:t, -100:, -100].flatten(), den_001[1:t+1, -100:, -100].flatten())
plt.title('denoised 001/scan_0001')
plt.xlabel('pixel value at time i')
plt.ylabel('pixel value at time i+1')
```
