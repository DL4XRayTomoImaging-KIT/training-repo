```python
from src.callbacks import iou
from univread import read as imread
```

```python
from glob import glob
from tqdm.auto import tqdm
import os
```

```python
def calculate_volumetric_error(prediction, label):
    if isinstance(prediction, str):
        prediction = imread(prediction)
    label = imread(label)
    
    return iou(preds=prediction, labels=label, C=[1,])
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
import numpy as np
```

```python
all_ious_biomedisa = []
all_ious_nearest = []
all_ious_nearest_5 = []
all_ious_nearest_15 = []
experiment_name = 'Task01_BrainTumour'

for label_addr in tqdm(glob(f'/mnt/data/decathlon/{experiment_name}/train/labels/*.tif')):
    filename = os.path.basename(label_addr)
    biomedisa_addr = os.path.join(f'/mnt/data/decathlon/{experiment_name}/train/labels_biomedisa/', filename)
    nearest_addr = os.path.join(f'/mnt/data/decathlon/{experiment_name}/train/labels_nearest/', filename)
    nearest_5_addr = os.path.join(f'/mnt/data/decathlon/{experiment_name}/train/labels_nearest_5/', filename)
    nearest_15_addr = os.path.join(f'/mnt/data/decathlon/{experiment_name}/train/labels_nearest_15/', filename)
    all_ious_biomedisa.append(calculate_volumetric_error(biomedisa_addr, label_addr))
    all_ious_nearest.append(calculate_volumetric_error(nearest_addr, label_addr))
    all_ious_nearest_5.append(calculate_volumetric_error(nearest_5_addr, label_addr))
    all_ious_nearest_15.append(calculate_volumetric_error(nearest_15_addr, label_addr))
    
print('biomedisa: ', np.median(all_ious_biomedisa))
print('nearest_neighbour 5: ', np.median(all_ious_nearest_5))
print('nearest_neighbour 10: ', np.median(all_ious_nearest))
print('nearest_neighbour 15: ', np.median(all_ious_nearest_15))

plt.figure(figsize=(16, 10))
sns.kdeplot(np.concatenate(all_ious_biomedisa), label='biomedisa');
sns.kdeplot(np.concatenate(all_ious_nearest), label='nearest neighbour 10');
sns.kdeplot(np.concatenate(all_ious_nearest_5), label='nearest neighbour 5');
sns.kdeplot(np.concatenate(all_ious_nearest_15), label='nearest neighbour 15');
plt.legend()
plt.show()
```
